## Random Forest Regression

1.Importing the Libraries

```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

2.Import Dataset File

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[ : ,1:-1].values
y = dataset.iloc[ : , -1].values
```

3.Training the Random Forest Regressor with the whole dataset

```python
from  sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor( n_estimators=10 , random_state=0 )
regressor.fit(X , y)
```

4.Predicting the new Result

```python
regressor.predict([[6.5]])
```

5.Visualizing the Random Forest Regerssion result

```python
min_val = X.min()
max_val = X.max()

X_grid = np.arange(min_val,max_val,0.01)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title("Random Forest Regression")
plt.xlabel('Positon Level')
plt.ylabel('Salary')
plt.show()
```
